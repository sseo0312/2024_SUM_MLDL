## **머신러닝 시작하기(간단한예제를 보자!)**

```python
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Y=2X-1 의 데이터

model=Sequential([Dense(units=1, input_shape=[1])])
#가장 간단한 신경망으로 하나의 layer와 하나의 neuron(동그라미)
#Dense란 한층의 모든 뉴런이 다른 층의 모든 뉴런과 연결된 상태
#input_shape입력 데이터의 크기, 이 경우 입력데이터는 X하나이므로 [1]로 지정

model.compile(optimizer='sgd', loss='mean_squared_error')
#sgd(경사하강법)을 사용해 추측한 값에서 오차(손실)을 계산->정답과 비교해 loss를 최소화하는 추측?을 하는 함수

xs=np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys=np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs,ys,epochs=500)
#X와 Y를 사용해 총 500번 반복해서 학습하라는 의미
#시간이 지남에따라 loss는 줄어들것임
#이는 컴퓨터가 둘 사이의 관계에 대해 반복을 여러번할수록 더 잘학습한다는 것을 의미

print(model.predict([10.0]))
#예측값 18.9
```

```python
print("신경망이 학습한것: {}".format(model.get_weights()))
#뉴런은 가중치(weight)와 절편(bias)를 학습한다
#이 경우 Y=WX+B라 나타낼 수 있고 이 모델은 정확히 Y=2X-1의 관계이다. (수학에서는 W를 기울기라 부르지만 머신러닝에서는 가중치라 부르는 경우가 더 많다)
#이 경우 W=1.9 B=-0.99정도로 학습함
```
